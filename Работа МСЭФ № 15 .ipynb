```python
# Получение данных об акциях с помощью yfinance
stock_data = yf.download('YNDX.ME', start='2021-01-01', end='2021-10-23')

# Рассчитываем  простую скользящую среднюю (SMA)
stock_data['SMA'] = stock_data['Close'].rolling(window=20).mean()

# Рассчитываем экспоненциальную скользящую среднюю (EMA)
stock_data['EMA'] = stock_data['Close'].ewm(span=20, adjust=False).mean()

# построим данные по акциям и скользящим средним
plt.figure(figsize=(12, 6))
plt.plot(stock_data['Close'], label='Close Price')
plt.plot(stock_data['SMA'], label='SMA')
plt.plot(stock_data['EMA'], label='EMA')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Analysis for YNDX')
plt.legend()
plt.show()
```

```python
!pip install yfinance    # установка библиотеки yfinance
!pip install matplotlib  # установка библиотеки matplotlib

import yfinance as yf
import matplotlib.pyplot as plt

# Получение данных об акциях Яндекса
stock = yf.Ticker("YNDX.ME")
data = stock.history(period="2y")  # получение данных за последние 2 года

# Построение графика цены акций
plt.figure(figsize=(12, 6))
plt.plot(data['Close'])
plt.title('Цена акций Яндекса')
plt.xlabel('Дата')
plt.ylabel('Цена акции')
plt.show()

# Анализ данных и прогнозирование
last_price = data['Close'][-1]
prev_price = data['Close'][-2]

if last_price > prev_price:
    decision = 'Покупать'
elif last_price < prev_price:
    decision = 'Продавать'
else:
    decision = 'Держать'

print(f'Цена акции на последнем торговом дне: {last_price}')
print(f'Решение: {decision}')
```
